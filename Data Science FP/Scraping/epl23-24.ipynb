```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup

# Setting up WebDriver 
driver = webdriver.Chrome()  
url = 'https://fbref.com/en/comps/9/2023-2024/stats/2023-2024-Premier-League-Stats'

# Opening the page with Selenium
driver.get(url)

# Get the page source after Cloudflare verification
soup = BeautifulSoup(driver.page_source, 'lxml')
driver.quit()


# Parse the table by the ID
tables = pd.read_html(str(soup), attrs={'id': 'stats_standard'})  
shooting_stats = tables[0] 


df = shooting_stats
```

```python
#renaming the columns
df = df.rename(columns={
    'Unnamed: 0_level_0': '',
    'Unnamed: 1_level_0': '',
    'Unnamed: 2_level_0': '',
    'Unnamed: 3_level_0': '',
    'Unnamed: 4_level_0': '',
    'Unnamed: 5_level_0': '',
    'Unnamed: 6_level_0': '',
    'Unnamed: 36_level_0': '',
})
df.head()
```

```python
#checking null values
df.info()
```

```python
#dropping the useless columns
df.drop(('', 'Matches'), axis=1, inplace=True)
```

```python
# Set pandas to display all rows
pd.set_option('display.max_rows', None)

# Inspecting all data from the df
df
```

```python
# dropping columns with no player data 
df.drop(index=[25, 51, 77, 103, 129, 155, 181, 207, 233, 259, 285, 311, 337, 363, 389, 415, 441, 467, 493, 519, 545, 571, 597], inplace=True)
```

```python
#checking for missing values
df.info()
```

```python
#dropping useless columns
df['','Nation'] = df['',"Nation"].str.split(' ', n=1).str[1]
```

```python
#inspecting df
df.head()
```

```python

# Defining the file path 
file_path = 'epl23-24 standard.csv'  

# Saving the DataFrame as a CSV file
df.to_csv(file_path, index=False)
```

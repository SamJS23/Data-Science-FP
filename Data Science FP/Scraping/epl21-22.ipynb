```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup

# Setting up the WebDriver
driver = webdriver.Chrome()  
url = 'https://fbref.com/en/comps/9/2021-2022/stats/2021-2022-Premier-League-Stats'

# Opening the page with Selenium
driver.get(url)

# Get the page source after Cloudflare verification
soup = BeautifulSoup(driver.page_source, 'lxml')
driver.quit()


# Parse the table by the ID 
tables = pd.read_html(str(soup), attrs={'id': 'stats_standard'})  
shooting_stats = tables[0] 


df = shooting_stats
df.head()
```

```python
# Renaming the columns
df = df.rename(columns={
    'Unnamed: 0_level_0': '',
    'Unnamed: 1_level_0': '',
    'Unnamed: 2_level_0': '',
    'Unnamed: 3_level_0': '',
    'Unnamed: 4_level_0': '',
    'Unnamed: 5_level_0': '',
    'Unnamed: 6_level_0': '',
    'Unnamed: 36_level_0': '',
})
df.head()
```

```python
#checking for null values
df.info()
```

```python
#dropping useless columns
df.drop(('', 'Matches'), axis=1, inplace=True)
```

```python

# Set pandas to display all rows
pd.set_option('display.max_rows', None)

# Display all rows
df
```

```python
#removing rows with no player data
df.drop(index=[25, 51, 77, 103, 129, 155, 181, 207, 233, 259, 285, 311, 337, 363, 389, 415, 441, 467, 493, 519, 545], inplace=True)
```

```python
#checking for null values again
df.info()
```

```python
#making the nations consistent 
df['','Nation'] = df['',"Nation"].str.split(' ', n=1).str[1]
```

```python
#inspecting the df
df.head()
```

```python


#defining the file path
file_path = 'epl21-22 standard.csv'  

# Saving the DataFrame as a CSV file
df.to_csv(file_path, index=False)
```
